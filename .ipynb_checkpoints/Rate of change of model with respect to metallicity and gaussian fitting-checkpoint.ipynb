# Analysis of $ \frac{dM}{dlnZ} $ and gaussian fitting
This notebook contains some initial analysis on the rate of change of the flux (model) with respect to the metalicity Z [Fe/H]. The python library lmfit was also used to approximate the data using gaussian fitting with the goal of gathering useful parameters of the curve, namely the wavelength where the flux is minimum (lambda0) and sigma.


In the code bellow I implement the function read_spec() which reads the data coming from a .fits file

````python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]

def read_spec(filename):
    hdul = fits.open(filename)
    flux = hdul[0].data
    header = hdul[0].header
    hdul.close()
    ln_wavelength = np.arange(header['NAXIS1'])*header['CDELT1']+header['WSTART']
    wavelength = np.exp(ln_wavelength)
                              
    return {'wavelength': wavelength, 'flux': flux, 'header': header}
                              
````

Using the following code it's possible to see the rate of change of the model with respect to the metallicity Z both for $T_{eff} = 5000 K$ and $T_{eff} = 5500 K$

````python
#derivative with respect to z (for T_eff=5000K)
k = 0.0
while(k>=-3.0):
    zi = "-"+str(k) if k==0.0 else str(k)
    zf= str(k-0.5)
    if k==-3.0:#jump from z=-3.0 to z=-4.0
        zf= str(k-1.0) 
        k-=1.0
    else:
        k-=0.5
    
    specf = read_spec('5000K/lte05000-1.50'+zf+'.PHOENIX-ACES-AGSS-COND-2011-HiRes.fits')
    speci = read_spec('5000K/lte05000-1.50'+zi+'.PHOENIX-ACES-AGSS-COND-2011-HiRes.fits')
    dM = specf['flux']-speci['flux']   
    dlnZ = -1.0 if k==-4.0 else -0.5 
    dSpec = dM/dlnZ 
    wav = specf['wavelength'] 
    plt.title('T=5000K, from Z='+zi+' to Z='+zf)
    plt.plot(wav, dSpec, linewidth=1)
    plt.xlabel('wavelength(Angstrom)')     
    plt.ylabel('dM/dZ')
    plt.plot(wav, [0 for i in range(len(wav))], linestyle='dashed') 

    plt.show() 
````

````python
#derivative with respect to z (for T_eff=5500K)
k = 0.0
while(k>=-3.0):
    zi = "-"+str(k) if k==0.0 else str(k)
    zf= str(k-0.5)
    if k==-3.0:#jump from z=-3.0 to z=-4.0
        zf= str(k-1.0) 
        k-=1.0
    else:
        k-=0.5
    specf = read_spec('5500K/lte05500-1.50'+zf+'.PHOENIX-ACES-AGSS-COND-2011-HiRes.fits')
    speci = read_spec('5500K/lte05500-1.50'+zi+'.PHOENIX-ACES-AGSS-COND-2011-HiRes.fits')
    dM = specf['flux']-speci['flux']   
    dlnZ = -1.0 if k==-4.0 else -0.5 
    dSpec = dM/dlnZ
    wav = specf['wavelength'] 
    plt.title('T=5500K, from Z='+zi+' to Z='+zf)
    plt.plot(wav, dSpec, linewidth=1)
    plt.xlabel('wavelength(Angstrom)')     
    plt.ylabel('dM/dZ')
    plt.plot(wav, [0 for i in range(len(wav))], linestyle='dashed') 
    #plt.savefig('5500K/derivative'+zi+'to'+zf+'.png', dpi=500)
    plt.show() 
````

Using the data for $Z=-2.5$ and $T_{eff} = 5000 K$ we have

````python
spec_gauss= read_spec('5000K/lte05000-1.50-2.5.PHOENIX-ACES-AGSS-COND-2011-HiRes.fits')
flux_gauss = spec_gauss['flux']
wave_gauss = spec_gauss['wavelength']

print("entire spectrum")
plt.plot(wave_gauss, flux_gauss, linewidth=1)
plt.title('T=5000K and Z=-2.5')
plt.xlabel('wavelength(Angstrom)')
plt.ylabel('Flux')
plt.show()


print("closeup on three distinct calcium absorption lines")
flux = flux_gauss[20200:21600]
wav = wave_gauss[20200:21600]
plt.plot(wav, flux)
plt.title('T=5000K and Z=-2.5')
plt.xlabel('wavelength(Angstrom)')
plt.ylabel('Flux')
plt.show()

````

Bellow I manually search for each absorption line in the region mentioned and, using the function fit_gaussian(), fit a gaussian that best represents the data.

````python
from numpy import exp, sqrt, pi
from lmfit import Model

def gaussian(x, amp, lambda0, sigma):
    return (amp / (sqrt(2*pi) * sigma)) * exp(-(x-lambda0)**2 / (2*sigma**2))

#function responsible to fit gaussion curve on data
def fit_gaussian(wavelength, model, amp, lambda0, sigma):
    gmodel = Model(gaussian)
    result = gmodel.fit(model-model[0], x=wavelength, amp=amp, lambda0=lambda0, sigma=sigma)#subtracting y0 so that best fit can start with y0=0
    plt.plot(x, y, 'o')
    plt.plot(x, result.best_fit+model[0], 'r-', label='best fit') #adding y0 to best fit data so that it can be shifted
    plt.legend(loc='best')
    plt.xlabel('wavelength(Angstrom)')
    plt.ylabel('Flux')
    plt.show()
    return result


#closeup on first distinct calcium absorption lines

a = 20445
b = 20470

y = flux_gauss[a:b]
x = wave_gauss[a:b]
result = fit_gaussian(x, y, 1,8497, 4)
print("Interval: ["+str(wave_gauss[a])+", "+str(wave_gauss[b])+"]")
print('lambda0: '+str(result.params['lambda0'].value)+'\n')

#closeup on second distinct calcium absorption lines

c = 20615
d = 20640

y = flux_gauss[c:d]
x = wave_gauss[c:d]
result2 = fit_gaussian(x, y, 1,8547, 4)
print("Interval: ["+str(wave_gauss[c])+", "+str(wave_gauss[d])+"]")
print('lambda0: '+str(result2.params['lambda0'].value)+'\n')

#closeup on last distinct calcium absorption lines
i = 21080
f = 21105

y = flux_gauss[i:f]
x = wave_gauss[i:f]
result3 = fit_gaussian(x, y, 1, 8655, 4)
print("Interval: ["+str(wave_gauss[i])+", "+str(wave_gauss[f])+"]")
print('lambda0: '+str(result3.params['lambda0'].value)+'\n')
````

````python
print("Hydrogen Beta absorption line")
i=1815
f=1870
x=wave_gauss[i:f]
y=flux_gauss[i:f]
result = fit_gaussian(x,y,1,4860,2)
print('Amplitude value: '+str(result.params['amp'].value))
print('Sigma value: '+str(result.params['sigma'].value))
print('Lambda0 value: '+str(result.params['lambda0'].value)+' Amgstrom')
print('Relative error lambda: '+str(100*abs(result.params['lambda0'].value-4861.3615)/486.13615)+' %')
````
